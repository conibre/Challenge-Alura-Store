# 4. Productos más y menos vendidos

###Gráfico N° 4 Producto Más y Menos Vendidos por Tienda



```python
def total_categoria(df):
    return df['Categoría del Producto'].value_counts().to_dict()

def producto_mas_vendido(pmv):
    categoria_valor = total_categoria(pmv)
    max_cantidad = -1
    categoria_popular = None
    for cat, cantidad in categoria_valor.items():
        if cantidad > max_cantidad:
            max_cantidad = cantidad
            categoria_popular = cat
    return [categoria_popular, max_cantidad]

def producto_menos_vendido(pmv):
    categoria_valor = total_categoria(pmv)
    min_cantidad = float('inf')
    categoria_menos_popular = None
    for cat, cantidad in categoria_valor.items():
        if cantidad < min_cantidad:
            min_cantidad = cantidad
            categoria_menos_popular = cat
    return [categoria_menos_popular, min_cantidad]

print(f'tienda:lo mas valorado es{producto__mas_vendido(tienda)} y lo menos valorados{producto_menos_vendido(tienda)}')
print(f'tienda2:lo mas valorado es{producto__mas_vendido(tienda2)} y lo menos valorados{producto_menos_vendido(tienda2)}')
print(f'tienda3:lo mas valorado es{producto__mas_vendido(tienda3)} y lo menos valorados{producto_menos_vendido(tienda3)}')
print(f'tienda4:lo mas valorado es{producto__mas_vendido(tienda4)} y lo menos valorados{producto_menos_vendido(tienda4)}')
```

```python
import matplotlib.pyplot as plt
import numpy as np


tiendas=[tienda, tienda2, tienda3, tienda4]
nombres_tiendas= ["tienda1", "tienda2","tienda3", "tienda4"]


mas_vendidos=[producto__mas_vendido(tienda)[1] for tienda in tiendas]

menos_vendido=[producto_menos_vendido(tienda)[1] for tienda in tiendas]


categoria_mas= [producto__mas_vendido(tienda)[0] for tienda in tiendas]
categoria_menos= [producto_menos_vendido(tienda)[0] for tienda in tiendas]

plt.figure(figsize=(10, 6))
bar_width=0.35
indices= np.arange(len(tiendas))


plt.bar(indices, mas_vendidos, bar_width, color= 'green', label= 'mas_vendidos')
plt.bar(indices + bar_width, menos_vendido, bar_width, color = 'red', label= 'menos_vendido')

plt.title('Producto Más y Menos Vendidos por Tienda')
plt.xlabel('Tienda')
plt.ylabel('cantidad de Productos')
plt.xticks(indices + bar_width/2, nombres_tiendas)

plt.legend()

for i in range(len(tiendas)):
    plt.text(indices[i], mas_vendidos[i]+ 5, categoria_mas[i], ha='center', fontsize=9)
    plt.text(indices[i] + bar_width, menos_vendido[i] + 5, categoria_menos[i], ha= 'center', fontsize=9)

plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
def total_envio(te):
  # Change 'Costo de envio' to the correct column name
  # For example, if the column name is 'Costo De Envío', then use:
  envios_total = te["Costo de envío"].sum()
  return envios_total

def promedio_envios(pe):
  envios_total = total_envio(pe)
  promedio_entrega = envios_total / len(pe)
  return round(promedio_entrega, 2)

print('total de envio de tienda:', total_envio(tienda))
print('promedio de envio de tienda:', promedio_envios(tienda))
print('total de envio de tienda2:', total_envio(tienda2))
print('promedio de envio de tienda:', promedio_envios(tienda2))
print('total de envio de tienda3:', total_envio(tienda3))
print('promedio de envio de tienda:', promedio_envios(tienda3))
print('total de envio de tienda4:', total_envio(tienda4))
print('promedio de envio de tienda:', promedio_envios(tienda4))
```
